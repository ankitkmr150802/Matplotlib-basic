# **Matplotlib**

**Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.  
It is one of the most widely used and powerful plotting libraries available in Python.  
Matplotlib allows you to create a wide range of plots, including line plots, scatter plots, bar plots, histograms, pie charts, and more.**.


```python
import matplotlib.pyplot as plt
```

# Line Plot

```python
# creating line plot

plt.plot([1,2,3],[2,5,7])
plt.show()
```

```python
#xlabel and ylabel

plt.plot([1,2,3],[2,5,7])
plt.title("Random data")
plt.xlabel("x-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 6, 8, 10]

plt.plot(x, y, label = "first")
plt.plot(x2, y2, label = "second")
plt.title("line comaprision")
plt.legend()  #to label
plt.show()
```

# Scatter plot

```python
#sample data

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x,y,marker = "*", label = "datapoint")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter-plot")

plt.grid(True)

#todisplay
plt.show()
```

```python
import numpy as np

x = np.random.rand(20)
y = np.random.rand(20)

color = np.random.rand(20)

plt.scatter(x, y, c= color)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter-plot")
plt.show()
```

# Bar Plot

```python
#sample data

categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 15, 25, 30]

plt.bar(categories,values)
plt.xlabel("categories")
plt.ylabel("values")
plt.title("Bar Plot")
plt.show()
```

```python
#Horizontal Bar Graph

categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 15, 25, 30]

plt.barh(categories,values)
plt.ylabel("categories")
plt.xlabel("values")
plt.title("Bar Plot")
plt.show()
```

```python
# grouped barplot

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [10, 20, 15, 25, 30]
values2 = [15, 25, 20, 30, 35]

width = 0.35
plt.bar(np.arange(len(categories)), values1, width = width, label='Group 1')
plt.bar(np.arange(len(categories)) + width, values2, width = width, label='Group 2')

plt.xlabel("categories")
plt.ylabel("values")
plt.title("Grouped chart")
plt.xticks(np.arange(len(categories)) + width / 2, categories)

plt.legend()
plt.show()
```

# Histogram

```python
#sample data

data = [1, 1, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 6, 7, 8, 8, 9]
n = len(data)
plt.hist(data, bins = n, edgecolor='black')
#bin means how many block to be formed?

plt.xlabel("data itemns")
plt.ylabel("frequency")
plt.title("histogram")
plt.show()
```
